### Data Preprocessing

```python
from sklearn.preprocessing import RobustScaler
```

#### target

```python
# seperating target feature
y = train_data[['MedHouseVal']]
y.head()
```

```python
# shape of target
y.shape
```

<br>
<br>

#### train data

```python
# seperating training data
X_train = train_data.drop(["id", "MedHouseVal"], axis=1)
X_train.head()
```

```python
# lookign at shape
X_train.shape
```

```python
# scaler for scaling the dataset
scaler = RobustScaler()
```

```python
# scaling the dataset
X_train_scaled = scaler.fit_transform(X_train)

# converting to dataframe
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_train_scaled.head()
```

<br>
<br>

#### test data

```python
# seperating test data
X_test = test_data.drop(["id"], axis=1)
X_test.head()
```

```python
# looking at shape
X_test.shape
```

```python
# scalling test data
X_test_scaled = scaler.transform(X_test)

# converting to dataframe
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
X_test_scaled.head()
```
