### Data collection

```python
# training data
x_train = pd.read_csv("./data/x_train_preprocessed.csv")
x_train.head()
```

```python
# test data
x_test = pd.read_csv("./data/x_test_preprocessed.csv")
x_test.head()
```

```python
# target
y = pd.read_csv("./data/y_preprocessed.csv")
y.head()
```

```python
# training data + target
X = pd.concat([x_train, y], axis=1)
X.head()
```

<br>
<br>
<br>

### Stacking

```python
import pycaret.regression as pc
```

```python
# setting up data
housing = pc.setup(data=X, target='MedHouseVal', numeric_features=list(x_train.columns), use_gpu=True)
```

```python
# compare performance of given 5 algorithms
top_5 = pc.compare_models(include=['lightgbm', 'catboost', 'xgboost', 'rf', 'et'], n_select=5)
```

```python
# stack top 5 algorithms
stacker = pc.stack_models(top_5)
```

<br>
<br>
<br>

### Creating submission

```python
# drop AveOccup from test set
test_set = x_test.drop('AveOccup', axis=1)
```

```python
# making predictions using our stacked models
predictions = stacker.predict(test_set)
```

```python
# inspecting predictions
predictions
```

```python
# creating submission dataframe
sub_df = pd.read_csv("./data/sample_submission.csv")
sub_df.head()
```

```python
# updating dummy values with actual predictions
sub_df.MedHouseVal = predictions
```

```python
# inspecting final dataframe
sub_df.head()
```

```python
# saving as a csv file
sub_df.to_csv("./data/submission_2.csv", index=None)
```
